# EDA: Synthetic Churn Dataset

Exploratory data analysis on the synthetic churn dataset generated by `src.data.make_dataset`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Add project root to PYTHONPATH
sys.path.insert(0, os.path.abspath('..'))
```

```python
from src.data.make_dataset import generate_synthetic_data

# Generate data (temporarily in notebooks/ or loaded from memory)
df = generate_synthetic_data("../data/raw/dataset.csv", n_samples=1000)

display(df.head())
display(df.describe())
```

```python
plt.figure(figsize=(6, 4))
df['target'].value_counts().plot(kind='bar')
plt.title('Class Balance')
plt.xlabel('Target (Churn)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```
